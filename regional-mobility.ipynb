```
import pandas as pd

mob = pd.read_csv('https://github.com/EvandroZanatta/ZanattaML/blob/master/regional-mobility.csv?raw=true')
```

```
mob.head()
```

```
# Get data only of Brazil
mob = mob[mob.country.str.contains('Brazil')]
mob = mob[mob.region.str.contains('Total')]
```

```
mob = mob.set_index('date')

mob.plot.line(
    figsize=(18,8), 
  )
```
